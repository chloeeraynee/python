```python
# importing a library under a callname
import pandas as pd
```

```python
# reading in a dataset
iris = pd.read_csv("iris.csv")
iris
```

```python
# using .describe() to summarize data
iris.describe()
```

```python
# selecting a data column from a dataset using the attribute name
iris['petal.length']
```

```python
# selecting a data column from a dataset using indicies
iris.iloc[:, 3]
```

```python
# selecting a specific entry attribute using indicies
print (iris['petal.length'][0])
print (iris['petal.length'][107])
```

```python
# selecting a small subset of entries and attributes using indicies
iris.iloc[[1,53,106], [1,2]]
```

```python
# selecting a subset of data entries with a specific criteria
iris.loc[iris['petal.width'] >= 2.0]
```

```python
# using MatPlotLib
import matplotlib.pyplot as plot
plot.scatter(iris['petal.length'], iris['petal.width'])
plot.title("Petal Length vs. Width")
plot.xlabel("Length")
plot.ylabel("Width")
```

```python
# using MatPlotLib on a subset of the data
iris = pd.read_csv("iris.csv")
setosa = iris.loc[iris['variety'] == 'Setosa']
plot.scatter(setosa['petal.length'], setosa['petal.width'])
plot.title('Setosa Petal Length vs. Width')
plot.xlabel('Length')
plot.ylabel('Width')
```
